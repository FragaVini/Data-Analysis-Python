```python
# Import libraries that will be used
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Importing the database
sport_toughest = pd.read_csv('toughestsport.csv', delimiter=',')
```

```python
# Printing the first rows
sport_toughest.head()
```

```python
# Checking data types and null values
sport_toughest.info()
```

```python
# Descriptive analysis of variables
sport_toughest.describe()
```

It is possible to identify that means between the skills are similar, with durability and power having the lowest and highest values respectively.

```python
# Create a new dataframe just with the skills to create the boxplot
sport_skills = sport_toughest.drop(['Total', 'Rank'], axis=1)
sport_skills.head()
```

```python
# Skill score variance
sport_toughest.var(axis=0, numeric_only=True)
```

```python
# Creating a boxplot of the skills
plt.figure(figsize=(10,6), dpi=600)
sns.boxplot(data=sport_skills, width = 0.6, palette='pastel')
plt.title('Skill Scores', fontsize=10)
plt.xlabel('Sport Skills', fontsize=6)
plt.ylabel('Score', fontsize=6)
plt.xticks(fontsize=5)
plt.yticks(fontsize=5)
plt.show()
```

1: It can be seen that most of the skills show a wide range of values, therefore some skills are crucial in some sports but not in others.

2: Endurance, Hand-eye coordination, nerve and analytical aplitude are positively skewed (median closer to the bottom quartile), thus higher frequency of high valued scores in these skills. Otherwise, strenght and speed are negatively skewed (median closer to the top quartile), so higher frequency of low valued scores.

3: Speed and nerve have the biggest box, thus they have the highest variablity of values.

4: The only skill that showed outliers was flexibility, with one value below the lower fence and values above the upper fence.

```python
# Checking the correlations between skills
corr_skills = sport_skills.corr(numeric_only=True)
corr_skills
```

```python
# Creating a heatmap to show the correlations between skills
plt.figure(figsize=(8,5), dpi=600)
sns.heatmap(data=corr_skills, annot=True, fmt='.2f', annot_kws={'size':8}, cbar_kws={'shrink':0.5}, cmap='Blues')
plt.title('Correlations between skills', fontsize=10)
plt.xticks(fontsize=6)
plt.yticks(fontsize=6)
plt.show()
```

The highest correlations were between the variables strength and power (0.85), hand-eye coordination and analytical aptitude (0.73) and strength and durability (0.71).

The other correlations were moderate or weak. Some correlations were negative, but with low values.

```python
# Calculate the p-value to check if the correlations are statistically significant
from scipy.stats import pearsonr

list_pvalue = []

list_pvalue.append(pearsonr(sport_skills['Strength'], sport_skills['Power'])[1])
list_pvalue.append(pearsonr(sport_skills['Hand-eye coordination'], sport_skills['Analytical Aptitude'])[1])
list_pvalue.append(pearsonr(sport_skills['Strength'], sport_skills['Durability'])[1])

list_pvalue
```

If we assign a significance level of 1%, the correlations are statistically significant, since the p-value was less than 0.01.

Thus, sports that require more power also require more force, which is understandable since one variable is dependent on the other, given that power depends on the amount of force applied and the speed at which that force is applied.

Agility, durability and speed are the three skills with the highest negative correlations, indicating that these skills are crucial to having a higher position in the ranking

```python
# Creating a dataframe without Total column to calculate the correlation between skills and rank
rank = sport_toughest.drop(['Total'], axis=1)
corr_rank = rank.corr(numeric_only=True)
corr_rank.iloc[0:10,-1].sort_values(ascending=True)
```

Agility, durability and speed are the three skills with the highest negative correlations with the ranking, which indicates that the higher the values of these variables, the better the position in the ranking. Therefore, the toughest sports are those that rely more on agility, durability and speed.

```python
# Creating a list with only the skills to use as a reference for ranking by skills
skills = sport_skills.columns.drop('SPORT')
skills.tolist()
```

```python
# Creating a dataframe with only the skills using the sports as index, and ranking them in descending order by the ranking of the original dataframe
ranked_sports = sport_toughest.set_index('SPORT')[skills].rank(ascending=False, axis=0)
# Create a loop through lamba to return only the first 5 values
top_five = ranked_sports.apply(lambda x : x.sort_values().index[:5])
top_five
```

The table above shows the 5 most difficult sports by skill, for example the sport that requires the most endurance is Cycling: Distance, and the sport that requires the most agility is Soccer.
