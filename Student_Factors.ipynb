#### Simple Linear Regression

```python
# Carry out a linear regression to predict the value of the exam
from sklearn.linear_model import LinearRegression
```

```python
# Create a regression only with the hours studied
X1 = new_df.Hours_Studied
y1 = new_df.Exam_Score
```

```python
# Creating the training and test data, with the size of the test data being 33% of the total
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X1, y1, test_size=0.33, random_state=42)
```

```python
# The value.reshape(-1,1) was used because there was only one column used in the regression
reg = LinearRegression().fit(X_train.values.reshape(-1, 1), y_train)
```

```python
# Importing methods to calculate prediction errors
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import r2_score
```

```python
# Evaluate the absolute error and squared error of the regression
y1_pred = reg.predict(X_test.values.reshape(-1, 1))

print(mean_absolute_error(y_test,y1_pred))
print(mean_squared_error(y_test,y1_pred))
```

```python
# Plotting the result of the prediction
fig,ax = plt.subplots()

ax.scatter(y1_pred,y_test)
ax.plot([60,75],[60,75], '--r')
plt.title('Linear Regression - Hours Studied', fontsize=10)
plt.xticks(fontsize=7)
plt.yticks(fontsize=7)

plt.show()
```

```python
# R-Squared to determine the proportion of variance
lf_r2 = r2_score(y_test, y1_pred)

print(f"R2 Score = {lf_r2}")
```

The value obtained is very low, so there is a wide range between the actual value and the predicted value.

```python
# Creating the training and test data, with the size of the test data being 33% of the total
X_train, X_test, y_train, y_test = train_test_split(X2, y2, test_size=0.33, random_state=42)
```

#### Removing Outliers

```python
# Establishing the first and third quartile to remove outliers
Q1 = new_df['Exam_Score'].quantile(0.25)
Q3 = new_df['Exam_Score'].quantile(0.75)
```

```python
# Calculating the interquartile range
IQR = Q3 - Q1
```

```python
# Formula to define low and high outliers
lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR
```

```python
# Creating a boolean array for low and high outliers
lower_array = np.where(new_df['Exam_Score'] <= lower)[0]
upper_array = np.where(new_df['Exam_Score'] >= upper)[0]
```

```python
# Resetting the index
new_df.reset_index(inplace=True)
```

```python
# Removing low outliers from the original dataset
new_df.drop(index=lower_array, inplace=True)
```

```python
# Removing high outliers from the original dataset
new_df.drop(index=upper_array, inplace=True)
```

```python
# Resetting the index
new_df.reset_index(inplace=True)
```

#### Multiple Linear Regression without Outliers

```python
# Creating a regression with all the data, but now without outliers
X_df = new_df.drop('Exam_Score',axis=1)
y_df = new_df.Exam_Score
```

```python
# Creating the training and test data, with the size of the test data being 33% of the total
X_train, X_test, y_train, y_test = train_test_split(X_df, y_df, test_size=0.33, random_state=42)
```

```python
reg = LinearRegression().fit(X_train, y_train)
```

```python
# Evaluate the absolute error and squared error of the regression
y_pred = reg.predict(X_test)

print(mean_absolute_error(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
```

```python
# Plotting the result of the prediction
fig,ax = plt.subplots()

ax.scatter(y_pred,y_test)
ax.plot([55,80],[55,80], '--r')
plt.title('Multiple Linear Regression - Without Outliers', fontsize=10)
plt.xticks(fontsize=7)
plt.yticks(fontsize=7)

plt.show()
```

```python
# R-Squared to determine the proportion of variance
df_r2 = r2_score(y_test, y_pred)

print(f"R2 Score = {df_r2}")
```

The value is good now, with good prediction, the only problem remains the same, it is not possible to detect the significance of this newly added variable.
