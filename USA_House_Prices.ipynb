# USA House Prices

```python
# Importing libraries that will be used
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Importing dataset
house_prices = pd.read_csv('USA Housing Dataset.csv', sep=',')
```

```python
# Plotting first rows
house_prices.head()
```

```python
# Plotting last rows to check for reference lines
house_prices.tail()
```

```python
# Checking datatype, shape and null values
house_prices.info()
```

```python
# Descriptive analysis of variables
house_prices.describe()
```

```python
# Changing the date column from string to date type
# First you need to create a dataset with only the date data
data = pd.to_datetime(house_prices['date']).dt.year
data.info()
```

```python
# Inserting the date data on the original dataset
house_prices.insert(0, 'year', data)
house_prices.head()
```

```python
# Removing the previous date column
house_prices.drop(['date'], axis=1, inplace=True)
```

```python
# Printing the column names
house_prices.columns
```

```python
# Establishing the first and third quartile to remove outliers
Q1 = house_prices['price'].quantile(0.25)
Q3 = house_prices['price'].quantile(0.75)
```

```python
# Calculating the interquartile range
IQR = Q3 - Q1
```

```python
# Formula to define low and high outliers
lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR
```

```python
# Creating a boolean array for low and high outliers
lower_array = np.where(house_prices['price'] <= lower)[0]
upper_array = np.where(house_prices['price'] >= upper)[0]
```

```python
# Removing low outliers from the original dataset
house_prices.drop(index=lower_array, inplace=True)
```

```python
# Removing high outliers from the original dataset
house_prices.drop(index=upper_array, inplace=True)
```

```python
# Resetting the index
house_prices.reset_index(inplace=True)
```

```python
# Checking the new shape
house_prices.shape
```

The previous dataset had 4140 rows, now the dataset has 3925, thus 215 was removed as outliers.

```python
# New descriptive analysis of variables
house_prices.describe()
```

The previous price mean was 5.530629e+05, and now the price mean is 4.820708e+05, which is a big difference.

```python
# Creating a dataset to make the correlation matrix
corr_house = house_prices.drop(['year'], axis=1)
```

```python
# Calculating the correlations between variables
corr_house = corr_house.corr(numeric_only=True)
```

```python
# Creating a heatmap to illustrate the correlation matrix
plt.figure(figsize=(15,9), dpi=600)
sns.heatmap(data=corr_house, annot=True, fmt='.2f', annot_kws={'size':8}, cbar_kws={'shrink':0.5}, cmap='coolwarm')
plt.title('Correlation Matrix', fontsize=10)
plt.xticks(fontsize=6)
plt.yticks(fontsize=6)
plt.show()
```

```python
# Creating a categorical variable to determine whether the house has been renovated or not
house_prices['renovated'] = np.where((house_prices['yr_renovated'] == 0), 'no',
                            np.where((house_prices['yr_renovated'] > 0), 'yes',
                                    'other'))
```

```python
# Printing the first 10 rows
house_prices.head(10)
```

```python
# Descriptive analysis of the variables based on grouping the data into renovated and non-renovated homes
renovated_stats = house_prices.groupby(['renovated'])
renovated_stats[['price','bedrooms','bathrooms','sqft_living','sqft_lot','condition']].describe().T
```

```python
# Calculating whether there is a significant difference between the means of the two sets of data
from scipy.stats import ttest_ind
group1 = house_prices[house_prices['renovated']=='no']
group2 = house_prices[house_prices['renovated']=='yes']

ttest_ind(group1['price'], group2['price'])
```

If we set a significance level of 1%, there is a significant difference between the means, so that non-renovated houses are significantly more expensive.

The problem is that the datasets are not the same size, with considerably more non-renovated houses than renovated ones.

```python
# Filtering the data that is above the 95th percentile, to evaluate the cities that have the most expensive houses
exp_city = house_prices.query('price > price.quantile(0.95)')
```

```python
# Checking the number of rows that have been selected
exp_city.shape
```

```python
# Making a plot to show the cities with the highest number of houses above the 95th percentile
plt.figure(figsize=(15,9), dpi=600)
ax = sns.countplot(y='city', hue='city', data=exp_city, palette='tab10')
# Plotting the numbers on the chart
for container in ax.containers: ax.bar_label(container, padding=1, fontsize=10)
plt.title('Most Expensive Cities')
plt.xlabel('Count')
plt.ylabel('City')
plt.show()
```

Based on the chart, you can see that Seattle and Bellevue are the cities with the highest number of expensive houses.
